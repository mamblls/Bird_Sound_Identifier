```python
# Usual Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn

# Librosa (the mother of audio files)
import os
import librosa
import librosa.display
import IPython.display as ipd
import warnings
warnings.filterwarnings('ignore')
```

```python
#this code is complex and i asked Co_piolet to help me out. I asked co-piolet to load my audio files and convert them into spectrograms. Display one per bird/file 
#then save all the spectrograms into the df called SGdf
#under each sectrogram it also printed the number of files per bird
#Due to the huge amount of computing power this was taking I decided to crop the sample size to  maxmium of 10 audio files per bird. and drop any that have less than 10 audio files
#In addition to this there were some files that were snagging the conversion process for a reason i couldn't figure out. So  i dropped them. like the Great tina
#this process took way to long and after 2 days of struggling with a dataset of 114 unique birds with 10-30 audio samples each i decided to cut it down to 10 audio samples per bird and nly 30 bird types

directory = 'Voice of Birds'
SGdf = pd.DataFrame()  # Initialize empty DataFrame to store spectrograms
dropped_birds = []  # Initialize list to store dropped birds
output_directory = 'Spectrograms'  # Directory to save spectrogram images

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop over all files in the directory - the files represent one species of bird and each contains multiple audio files
for foldername in os.listdir(directory):
    if os.path.isdir(os.path.join(directory, foldername)):
        count = 0  # Initialize count for each folder
        for filename in os.listdir(os.path.join(directory, foldername)):
            # Check if the file is an audio file - these are saved as mp3 files
            if filename.endswith('.mp3'):
                filepath = os.path.join(directory, foldername, filename)
                # Load the audio files
                audio, sr = librosa.load(filepath, sr=None)

                # Compute the Mel spectrogram
                S = librosa.feature.melspectrogram(y=audio, sr=sr)
                # Convert to log scale (dB)
                D = librosa.power_to_db(S, ref=np.max)

                # Add the spectrogram to the DataFrame
                SGdf = pd.concat([SGdf, pd.DataFrame(D.flatten(), columns=[foldername])], axis=1)

                count += 1  # Increment count for each audio file

                # Plot the Mel spectrogram
                plt.figure(figsize=(10, 4))
                librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='mel')
                plt.colorbar(format='%+2.0f dB')
                plt.title('Mel spectrogram: ' + filename)  # Include the filename in the title
                plt.tight_layout()

                # Save the figure as a PNG file
                plt.savefig(os.path.join(output_directory, filename + '.png'))

                # Close the figure to free up memory
                plt.close()

            # Break the loop if count reaches 10
            if count >= 10:
                break

        # If count is less than 10, drop the bird from the DataFrame and add it to the dropped birds list
        if count < 10:
            SGdf = SGdf.drop(columns=[foldername])
            dropped_birds.append(foldername)
        else:
            print(f'Bird: {foldername}, Number of audio files: {count}')  # Printing the name of each folder (also name of each bird) along with number of audio files in it

# Print the dropped birds
print('Dropped birds:', dropped_birds)
```
